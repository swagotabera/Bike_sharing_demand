## Mapping data

```python
#Mapping data with appropriate values
season_map={1:"spring",2:"summer",3:"fall",4:"winter"}
bike_demand_dataframe['season'] = bike_demand_dataframe['season'].map(season_map)

month_map={1:"Jan",2:"Feb",3:"Mar",4:"Apr",5:"May",6:"June",7:"July",8:"Aug",9:"Sep",10:"Oct",11:"Nov",12:"Dec"}
bike_demand_dataframe['mnth'] = bike_demand_dataframe['mnth'].map(month_map)

weekday_map={0:"Sunday",1:"Monday",2:"Tuesday",3:"Wednesday",4:"Thursday",5:"Friday",6:"Saturday"}
bike_demand_dataframe['weekday'] = bike_demand_dataframe['weekday'].map(weekday_map)

wheather_map={1:"Clear, Few clouds",2:"Mist + Cloudy",3:"Light Snow, Light Rain",4:"Heavy Rain + Ice Pallets"}
bike_demand_dataframe['weathersit'] = bike_demand_dataframe['weathersit'].map(wheather_map)
```

```python
bike_demand_dataframe.head()
```

## Dropping columns

```python
#Dropping instant column because it same as index
bike_demand_dataframe.drop(['instant'],axis=1,inplace=True)

#Dropping casual & registered column because both present in cnt column
bike_demand_dataframe.drop(['casual','registered'],axis=1,inplace=True)

#Dropping dteday column because yr & mnt column present
bike_demand_dataframe.drop(['dteday'],axis=1,inplace=True)

```

```python
#convert variable to object type
bike_demand_dataframe['mnth']=bike_demand_dataframe['mnth'].astype(object)
bike_demand_dataframe['season']=bike_demand_dataframe['season'].astype(object)
bike_demand_dataframe['weathersit']=bike_demand_dataframe['weathersit'].astype(object)
bike_demand_dataframe['weekday']=bike_demand_dataframe['weekday'].astype(object)
bike_demand_dataframe.info()
```

```python
#CREATING DUMMY VARIABLES FOR CATEGORICAL DATA

Season_condition=pd.get_dummies(bike_demand_dataframe['season'],drop_first=True)
bike_demand_dataframe=pd.concat([bike_demand_dataframe,Season_condition],axis=1)
Weather_condition=pd.get_dummies(bike_demand_dataframe['weathersit'],drop_first=True)
bike_demand_dataframe=pd.concat([bike_demand_dataframe,Weather_condition],axis=1)
Day_of_week=pd.get_dummies(bike_demand_dataframe['weekday'],drop_first=True)
bike_demand_dataframe=pd.concat([bike_demand_dataframe,Day_of_week],axis=1)
Month=pd.get_dummies(bike_demand_dataframe['mnth'],drop_first=True)
bike_demand_dataframe=pd.concat([bike_demand_dataframe,Month],axis=1)
```

```python
bike_demand_dataframe.head()
```

```python
#delete columns
bike_demand_dataframe.drop(['season'],axis=1,inplace=True)
bike_demand_dataframe.drop(['weathersit'],axis=1,inplace=True)
bike_demand_dataframe.drop(['weekday'],axis=1,inplace=True)
bike_demand_dataframe.drop(['mnth'],axis=1,inplace=True)
```

```python
bike_demand_dataframe_train, bike_demand_dataframe_test = train_test_split(bike_demand_dataframe, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
bike_demand_dataframe_test.head()
```

```python
#SCALING THE NUMERICAL DATA
scaler=MinMaxScaler()

num_vars=['temp','hum','windspeed']
bike_demand_dataframe_train[num_vars] = scaler.fit_transform(bike_demand_dataframe_train[num_vars])
num_vars_test=['temp','hum','windspeed']
bike_demand_dataframe_test[num_vars_test] = scaler.fit_transform(bike_demand_dataframe_test[num_vars_test])
```

```python
bike_demand_dataframe_train.head()
```

```python
#copying original value
bike_demand_dataframe_train_original=bike_demand_dataframe_train
bike_demand_dataframe_test_original=bike_demand_dataframe_test
```

```python
#CREATING X AND Y
y_train = bike_demand_dataframe_train.pop('cnt')
X_train = bike_demand_dataframe_train
X_test = bike_demand_dataframe_test
y_test = bike_demand_dataframe_test.pop('cnt')
```

```python
#creating copy of X & y test train
y_train_original=y_train
X_train_original=X_train
X_test_original=X_test
y_test_original=y_test
```

```python
#linear regressor model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred_test = regressor.predict(X_test)
y_pred_train = regressor.predict(X_train)
```
